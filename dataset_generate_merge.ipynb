```python
import nltk
from nltk import word_tokenize
from nltk.util import ngrams
from collections import Counter
import os
from csv import writer 
import csv
```

```python
nltk.download('punkt')
```

```python
#text=("Adduser.txt","r")
#give input folder that contains all merged files.
folder = "/home/2019aim1006/Downloads/Merged/"
```

```python
# Enter the value of n for n grams
n = 2
```

```python
#Enter value (m) for taking most frequent elements from n grams
m = 80
```

```python
feature_set = set()
```

```python
for fname in os.listdir(folder):
    curr_fname = folder+fname
    with open(curr_fname,'r') as file:
        text = file.read()
        token = nltk.word_tokenize(text)

        n_grams = ngrams(token,n)
        
        for f, count in Counter(n_grams).most_common(m):
            feature_set.add(f)
            #print(f)
```

```python

fields = list(feature_set)
    
# name of csv file  
filename = "/home/2019aim1006/Downloads/adfa_train"
filename+=str(n)
filename+=str(m)
filename+=".csv"
    
# writing to csv file  
with open(filename, 'w') as csvfile:  
    # creating a csv dict writer object  
    writer = csv.DictWriter(csvfile, fieldnames = fields)  
    
    #print(currfilename)
    # writing headers (field names)  
    writer.writeheader()  
        
```
